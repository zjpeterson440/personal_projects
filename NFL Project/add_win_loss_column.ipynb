```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv('qbs_for_viz.csv')
```

```python
#Converts results into 1/0 win loss values
result=[]
for i in range(len(weather)):
    if weather['Result'].iloc[i][0]=='W':
        result.append(1)
    else:
        result.append(0)
        
weather['win_loss'] = result
```

```python
margin_victory = []
for i in range(len(weather)):
    results = list(map(int, weather['Result'].iloc[i].split().pop(1).split('-')))
    score_dif = results[0] - results[1]
    margin_victory.append(score_dif)
weather['margin_victory'] = margin_victory
```

```python
weather.drop(['DewPoint', 'Pressure'], axis=1, inplace=True)
```

```python
weather.to_csv('qb_analysis_winloss_marginvictory.csv',index=False)
```
